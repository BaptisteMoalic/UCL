```python
def load_iris_dataset(num_classes=2):
    # Load the datadet from SkDataset
    iris = skdataset.load_iris()
    X = iris.data
    Y = iris.target
    # Reduce the number of classes
    idx = Y < num_classes
    X = X[idx]
    Y = Y[idx]
    return X, Y
```

```python
def load_mnist_dataset(num_classes=2):
    # Load the datadet from SkDataset
    X, Y = skdataset.fetch_openml('mnist_784', version=1, return_X_y=True,as_frame=False)
    Y = Y.astype(np.int64)
    # Reduce the number of classes
    idx = Y < num_classes
    X = X[idx]
    Y = Y[idx]
    return X, Y

def load_mnist_dataset_onevsall(class_id=7):
    # Load the datadet from SkDataset
    X, Y = skdataset.fetch_openml('mnist_784', version=1, return_X_y=True,as_frame=False)
    Y = Y.astype(np.int64)
    # One versus all
    idx = Y == class_id
    Y[~idx] = 0
    Y[idx]  = 1
    return X, Y
```

```python
def true_positive(Y_test, y_pred):
    mask = (Y_test == 1)
    tp = (Y_test[mask] == y_pred[mask]).sum()
    return tp.item()

def true_negative(Y_test, y_pred):
    mask = (Y_test == 0) | (Y_test == -1)
    tn = (Y_test[mask] == y_pred[mask]).sum()
    return tn.item()

def false_negative(Y_test, y_pred):
    mask = (y_pred == 0) | (y_pred == -1)
    tn = (Y_test[mask] != y_pred[mask]).sum()
    return tn.item()

def false_positive(Y_test, y_pred):
    mask = (y_pred == 1)
    tn = (Y_test[mask] != y_pred[mask]).sum()
    return tn.item()

def plot_confusion_matrix(Y_test, y_pred):
    tp = true_positive(Y_test, y_pred)
    tn = true_negative(Y_test, y_pred)
    fp = false_positive(Y_test, y_pred)
    fn = false_negative(Y_test, y_pred)
    cf = np.array([[tn, fp], [fn, tp]])
    fig, ax = plt.subplots()
    ax.matshow(cf, cmap=plt.cm.Blues)
    for i in range(2):
        for j in range(2):
            c = cf[i,j]
            ax.text(j, i, str(c), va='center', ha='center')
    plt.xlabel('Prediction')
    plt.ylabel('Target')
    plt.show()
```

```python
def nsolve(X,y):
    """
    Write your implementation of nsolve here.

    Arguments:
    X : Data matrix
    y : Labels

    Returns:
    w : Weights 
    """

    #begin_solution
    
    w = np.linalg.inv(X.T @ X) @ X.T @ y #This is the normal equation
    w = np.reshape(w, (1, -1))

    #end_solution


    return w
```

```python
def line_fit(X,y):
    """
    Write your implementation of line_fit here.

    Arguments:
    X : Data matrix
    y : Labels

    Returns:
    w : Weights
    l2_error : L2 Prediction error using learned w
    """

    #begin_solution
    
    w = nsolve(X, y) #We start by computing the weights for the regression

    y_pred = w @ X.T #We compute the linear estimation of the labels using the weights previously computed
    
    N = np.size(y_pred)
    
    l2_error = ((np.linalg.norm(y_pred-y))**2)/N #The L2 error is the Mean Squared Error
    
    #end_solution


    return w, l2_error
```

```python
def gradient_descent(parameters, objective, d_objective, step_size, max_iterations, eps_change=1.0e-6):
    """
    Write your own implementation of gradient descent here.
    
    Arguments:
    
    parameters:      init of parameters to optimize - R^N
    objective:       function to minimize - R^N -> R^1
    d_objective:     derivative of the function to minimize - R^N -> R^N
    step_size:       step size/learning rate for the optimization algorithm - R^1
    max_iterations:  maximum number of iterations to run the optimization algorithm - R^1 integer
    eps_change:      minimum change of minimizer one step to the next before stopping - R^1
    
    Return:
    minimizer:    solution of the optimization
    current_it:   iteration number at which the optimization reached the minimum
    """
    
    N          = parameters.shape[0]
    minimizer  = parameters.copy()
    current_it = 0
    
    #begin_solution
    
    while (current_it <= max_iterations):
        temp = minimizer #We keep the current minimizer so that we can compare it with the next one
        minimizer = minimizer - step_size*d_objective(minimizer) #Gradient descent formula
        if(np.sum(abs(temp-minimizer))<eps_change): #If we get stuck on a x (i.e its difference between two steps is < to epsilon)
            break #We break out of the loop
        current_it += 1 #Iterating
        
    #end_solution

        
    return minimizer, current_it + 1
    
```

```python
def gradient_descent_nesterov(parameters, objective, d_objective, momentum, step_size, max_iterations, eps_change):
    """
    Write your own implementation of gradient descent with Nesterov momentum here.
    
    Arguments:
    parameters:      init of parameters to optimize - R^N
    objective:       function to minimize - R^N -> R^1
    d_objective:     derivative of the function to minimize - R^N -> R^N
    momentum:        weight of the momentum term - R^1
    step_size:       step size/learning rate for the optimization algorithm - R^1
    max_iterations:  maximum number of iterations to run the optimization algorithm - R^1 integer
    eps_change:      minimum change of minimizer one step to the next before stopping - R^1
    
    Return:
    minimizer:    solution of the optimization
    current_it:   iteration number at which the optimization reached the minimum
    """
    
    N          = parameters.shape[0]
    minimizer  = parameters.copy()
    current_it = 0
    
    #begin_solution
    w = minimizer #Initialization
    v = minimizer #Initialization
    
    while (current_it <= max_iterations):
        temp = minimizer #We keep the current minimizer so that we can compare it with the next one
        v = momentum*v - step_size*d_objective(w + momentum*v) #Nesterov's GD formula
        w = w + v
        minimizer = w
        if(np.sum(abs(temp-minimizer))<eps_change): #If we get stuck on a x (i.e its difference between two steps is < to epsilon)
            break #We break out of the loop
        current_it += 1 #Iterating

    #end_solution

        
    return minimizer, current_it + 1
    
```

```python
def second_order(parameters, objective, d_objective, dd_objective, step_size, max_iterations, eps_change):
    """
    Write your own implementation of a second order optimizer.
    
    Arguments:
    parameters:      init of parameters to optimize - R^N
    objective:       function to minimize - R^N -> R^1
    d_objective:     function computing the derivative of the objective - R^N -> R^N
    dd_objective:    function computing the second derivative of the objective - R^N -> R^NxN
    step_size:       step size/learning rate for the optimization algorithm - R^1
    max_iterations:  maximum number of iterations to run the optimization algorithm - R^1 integer
    eps_change:      minimum change of minimizer one step to the next before stopping - R^1
    
    Return:
    minimizer:    solution of the optimization
    current_it:   iteration number at which the optimization reached the minimum
    """
    
    N          = parameters.shape[0]
    minimizer  = parameters.copy()
    current_it = 0
    
    #begin_solution
    
    while (current_it <= max_iterations):
        temp = minimizer #We keep the current minimizer so that we can compare it with the next one
        minimizer = minimizer - (d_objective(minimizer) @ np.linalg.inv(dd_objective(minimizer))) #Newton's formula
        if(np.sum(abs(temp-minimizer))<eps_change): #If we get stuck on a x (i.e its difference between two steps is < to epsilon)
            break #We break out of the loop
        current_it += 1 #Iterating  
    
    #end_solution

    return minimizer, current_it + 1
    
```

```python
def paraboloid(X): # paraboloid or saddle function X[0]^2-X[1]^2+X[2]^2 ... 
    tmp = np.power(X, 2)
    return tmp[0::2].sum() - tmp[1::2].sum()

def d_paraboloid(X):
    gradient = np.zeros(X.shape[0])
    
    #begin_solution
    gradient[0::2] = 2*X[0::2] #The derivative of x -> x^2 being x -> 2x
    gradient[1::2] = -2*X[1::2] #The derivative of x -> -x^2 being x -> -2x
    #end_solution

    return gradient

def dd_paraboloid(X):
    hessian = np.eye(X.shape[0])
    #begin_solution
    n = X.shape[0]
    for i in range(n): #Iterating on the lines
        for j in range(n): #Iterating on the columns
            if(i==j): #If i==j, it is the second derivative of x -> +/-x^2, thus x -> +/-2
                hessian[i, j] = ((-1)**j)*2
            #Otherwise, we derive by x_i, thus only having a x_i component, so after deriving by x_j, the result is 0
    #end_solution

    return hessian
```

```python
def pca(X, ndims):
    """
    Write your implementation of PCA here.

    Arguments:
    X : Data matrix
    ndims : Number of reduced dimensions

    Returns:
    mean_vec : Data mean
    top_eig_vecs : Selected eigen vectors, a matrix where each column corresponds
                    to an eigen vector
    """

    #begin_solution
    
    #X ROW=OBS, COLUMN=FEATURE
    
    #Computing the mean vector
    mean_vec = np.reshape(np.mean(X, axis=0), (-1, 1))
    
    #Centring the data
    ones_vec = np.reshape(np.ones(X.shape[0]), (-1, 1)) #Vector of ones to repeat the data on every observation
    X_c = X - (ones_vec @ mean_vec.T)
    
    #Computing the covariance matrix
    cov_matrix = np.cov(X_c.T) #ROW=FEATURE, COLUMN=OBSERVATION (inverted wrt to our X)
     
    #Computing the eigenvectors
    eig_val, eig_vecs = np.linalg.eig(cov_matrix)
    
    #Looking for only the ndims TOP components
    top_eig_vecs = np.zeros((eig_vecs.shape[0], ndims))
    for k in range(ndims):
        top_index = np.argmax(eig_val) #We look for the top eigenvalue
        top_eig_vecs[:, k] = eig_vecs[:, top_index] #We retrieve the corresponding eigenvector
        eig_val[top_index] = float('-inf') #We put this eigenvalue to minus infinity so it doesn't come up again
    
    #end_solution


    return mean_vec, top_eig_vecs
```

```python
def pca_projection(X, mean_vec, eig_vecs):
    """
    Write your implenetation of PCA projection here.

    Arguments:
    X : Data matrix
    mean_vec : Data mean
    eig_vecs : A numpy array where each column corresponds to an eigen vector

    Returns:
    weights : Weights corresponding to the eigen vectors
    """

    #begin_solution
    
    #Centring the data
    ones_vec = np.reshape(np.ones(X.shape[0]), (-1, 1)) #Vector of ones to repeat the data on every observation
    X_c = X - (ones_vec @ mean_vec.T)
    
    #Projection
    weights = X_c @ eig_vecs

    #end_solution


    return weights
```

```python
def pca_reconstruction(weights, mean_vec, eig_vecs):
    """
    Write your implementation of PCA reconstruction here.

    Arguments:
    weights : Weights obtained from pca_projection
    mean_vec : Data mean
    eig_vecs : A numpy array where each column corresponds to an eigen vector
    """

    #begin_solution
    
    #Projecting a second time
    reconstruction = weights @ eig_vecs.T
    
    #Adding back the mean value
    ones_vec = np.reshape(np.ones(reconstruction.shape[0]), (-1, 1)) #Vector of ones to repeat the data on every observation
    reconstruction += ones_vec @ mean_vec.T

    #end_solution


    return reconstruction
```

```python
def kmeans(X, k, max_iterations=1000):
    """
    Write your implementation of kmeans here.

    Arguments:
    X : Data matrix.
    k : Your choice of k, an integer.
    max_iterations : The number of iterations after which to stop.

    Returns:
    centroids: The learned cluster centroids.
    """

    #begin_solution
    
    # Write your code here:
    epsilon = 1e-4
    
    n, ndims = X.shape # Collecting the shape of the input data matrix
    
    clusters = np.random.uniform(0.0, 1.0, (k, ndims)) #Initializing the clusters to k number of images, with random pixels between 0 and 1
    
    L = np.zeros(n, dtype=int) #L keeps in memory to which mean we have assigned each point
    
    previous_clusters = clusters
    
    # First loop: we assign the x's to the different means
    for pt_idx in range(n):
        pt_value = X[pt_idx, :] # Iterating on the x's
        best_distance = np.linalg.norm(pt_value-clusters[0, :]) # We do as if the first centre was ok, given that L is initialized as only zeroes
        for t in range(1, k): # Iterating on the means
            if(np.linalg.norm(pt_value-clusters[t, :]) < best_distance): # If the distance to this mean is less than the best one, we update
                L[pt_idx] = t
                best_distance = np.linalg.norm(pt_value-clusters[t, :])
            
    # Second loop: we update the mean
    sum_values = np.zeros_like(clusters)
    nb_of_points = np.zeros(k)
    for pt_idx in range(n):
        sum_values[L[pt_idx], :] += X[pt_idx, :]
        nb_of_points[L[pt_idx]] += 1 
    for t in range(1, k):
        clusters[t, :] = sum_values[t, :]/(nb_of_points[t]+epsilon) # POSSIBLE DIVISION BY 0
            
    #While loop: max iterations
    nb_of_iterations = 0
    while(nb_of_iterations < max_iterations):
        
        #We break out of the loop if the centroids stay too similar between two iterations
        if(np.linalg.norm(clusters-previous_clusters)<epsilon):
            break
        
        #We stock the clusters we have, so as to compute new ones
        previous_clusters = clusters
        
        # First loop: we assign the x's to the different means
        for pt_idx in range(n):
            pt_value = X[pt_idx, :] # Iterating on the x's
            prev_distance = np.linalg.norm(pt_value-clusters[0, :]) # We do as if the first centre was ok, given that L is initialized as only zeroes
            for t in range(1, k): # Iterating on the means
                if(np.linalg.norm(pt_value-clusters[t, :]) < prev_distance): # If the distance to this mean is less than the best one, we update
                    L[pt_idx] = t
                    prev_distance = np.linalg.norm(pt_value-clusters[t, :])
            
        # Second loop: we update the mean
        sum_values = np.zeros_like(clusters)
        nb_of_points = np.zeros(k)
        for pt_idx in range(n):
            sum_values[L[pt_idx], :] += X[pt_idx, :]
            nb_of_points[L[pt_idx]] += 1 
        for t in range(1, k):
            clusters[t, :] = sum_values[t, :]/(nb_of_points[t]+epsilon)
        
        nb_of_iterations += 1
    
    centroids = clusters

    #end_solution


    return centroids
```

_Your reply_: The k-means algorithm initializes random centroids. We then assign each of our samples to these centroids based on similarity (here, the minimum L2-distance), and we re-compute the new centroids' coordinates so that we can distribute each sample between the k centroids (in the end, each centroid will represent the mean of a precise population within all of the samples).
The difference between PCA and k-means is that PCA is meant for dimensionality reduction (for compression, reconstruction), and k-means is meant for clustering. They are not meant for the same task, so it is difficult to compare them. k-means uses mainly the mean of a population of sample, whereas PCA uses the eigen-values/vectors to reduce the number of features.
We could maybe extract the eigen-vectors from a population of samples, and then compare how much each vector explains each sample, and cluster based on each eigen-vector.

k-means is not meant for reconstruction, but if we re talking about what each centroid represent as an image, we can say that in the end, it is hard to identify distinct results using k-means. There are multiple causes, including the fact that we initialize at random, and so there is a possibility that the random centroid is never assigned to any sample. Also, it is possible that two centroids get close to each other, and so represent the same feature. In the end, as our population is big, and that the result is only a mean, it is possible to have different contributions on each feature (0 and 8 looking alike, they might be assigned to the same centroid, and therefore the centroid will neither look like an 0 nor a 8, but something much more blurry).

```python
def lclass(examplesA, examplesB, testExample):
    """
    Write your implementation of normal equation based linear classifier here.

    Arguments:
    examplesA : Matrix of samples from class A
    examplesB : Matrix of samples from class B
    testExample : Matrix of test samples to predict on

    Returns:
    preds : Predicted 0/1 labels on testExample
    """

    #begin_solution
    nDataA = examplesA.shape[0]
    nDataB = examplesB.shape[0]
    nDataExamples = testExample.shape[0]
    
    # We say that the examples from A have the label 1, and from B -1
    y = np.ones((nDataA, 1))
    y = np.append(y, -np.ones((nDataB, 1)))
    
    # Adding bias
    examplesA = np.c_[np.ones((nDataA, 1)), examplesA]
    examplesB = np.c_[np.ones((nDataB, 1)), examplesB]
    testExample = np.c_[np.ones((nDataExamples, 1)), testExample]
    
    # Concatenating A and B matrices for examples
    X = np.concatenate((examplesA, examplesB))
    
    # Normal equation
    w = (np.linalg.inv(X.T @ X) @ X.T @ y).reshape((-1, 1))
    
    # Predicted labels
    preds = testExample @ w
    
    # If the result is negative => -1, positive => 0
    for k in range(preds.shape[0]):
        if(preds[k, :] >= 0):
            preds[k, :] = 0
        else:
            preds[k, :] = 1
    
    #end_solution


    return preds.reshape(-1)
```

```python
def lmclass(examples, labels, testExample):
    """
    Write your implementation of multiclass classifier lmclass here.
    Use your lclass_prob implementation here to solve this.

    Arguments:
    examples : The training data matrix
    labels : The training labels
    testExample : The testing data matrix

    Returns:
    preds_prob : For each sample in testExample, their predicted probabilities
                 of belonging to each class
    """

    #begin_solution
    
    # How many different labels are there?
    K = np.amax(labels)
    nData, nDim = testExample.shape
    preds_prob = np.zeros((nData, K+1))
    
    # We opt for the One VS Rest method: we will run the linear classifier on exampleA and exampleB where exampleA is the
    # set with samples of SAME LABEL, and exampleB the rest of the dataset
    # We obtain scores for each classifier, and we apply softmax on it to have probabilities
    
    for k in range(K): # For each label
        
        examplesA = np.zeros((1, nDim)).reshape((1, -1)) # Creating arrays with one element to have the right dimensions
        examplesB = np.zeros((1, nDim)).reshape((1, -1))
        for cData in range(nData):
            if(labels[cData] == k):
                examplesA = np.r_[examplesA, examples[cData, :].reshape((1, -1))] # We sort each value either in exampleA or B
            else:
                examplesB = np.r_[examplesB, examples[cData, :].reshape((1, -1))]
        examplesA = examplesA[1:, :] # We remove the first element we used for dimensioning
        examplesB = examplesB[1:, :]
        
        preds_k = lclass_prob(examplesA, examplesB, testExample) # These are the scores we obtain
        preds_prob[:, k] = preds_k.reshape(-1) # We store them in the corresponding column 
    
    # Once we have finished, we just have to "normalize" the rows using softmax
    for cData in range(nData):
        preds_prob[cData, :] = softmax(preds_prob[cData, :])

    #end_solution

    
    return preds_prob
```

```python
def logistic_loss(preds, targets):
    """
    Implement your logistic loss function here.

    Arguments:
    preds : Predicted values
    targets : Target values

    Returns :
    cost : The mean logistic loss value between preds and targets
    """

    #begin_solution
    # Peut-être vectorialiser ici
    epsilon = 1e-6
    Loss = 0
    K = preds.shape[0]
    
    for k in range(K):
        Loss += -targets[k] * np.log(sigmoid(preds[k]) + epsilon) - (1 - targets[k]) * np.log(1 + epsilon - sigmoid(preds[k]))
    
    Loss = Loss/K
    cost = Loss
    #end_solution


    return cost
```

```python
def dlogistic(preds, X, Y,W=[]):
    """
    Implement the derivative of the logistic loss here.

    Arguments:
    preds : Predicted values
    X : Input data matrix
    Y : True target values
    W : The weights, optional argument, may/may not be needed depending on the loss function
    """

    #begin_solution
    J = np.zeros((W.shape[0], 1))
    K = X.shape[0]
    
    for k in range(K):
        J += -X[k, :].T.reshape((-1, 1)) * (Y[k] - sigmoid(preds[k]))
        
    J = J/K
    J = np.reshape(J, (-1, 1))
    #end_solution

    
    return J
```

```python
def hinge_loss(z, Y):
    """
    Write your implementation of hinge loss here.

    Arguments:
    z : Classifier score
    Y : Label

    Returns:
    avg : The average hinge loss
    """

    #begin_solution
    Loss = 0
    K = Y.shape[0]
    z = np.reshape(z, -1)
    Y = np.reshape(Y, -1)
    print(z)
    print(Y)
    
    for k in range(K):
        Loss += max(0, 1 - z[k] * Y[k])
    
    Loss = Loss/K
    avg = Loss
    #end_solution


    return avg
  
def dhinge(preds,X, Y, W=[]):
    """
    Implement the derivative of hinge_loss here.

    Arguments:
    preds : The predicted values.
    X : Data matrix
    Y : True Labels
    W : Weights. Optional argument, may/may not be needed, depending on loss function
    Returns:
    J : Gradient of avg hinge_loss with respect to predicted values.
    """

    #begin_solution
    J = np.zeros((W.shape[0], 1))
    K = preds.shape[0]
    
    for k in range(K):
        if((Y[k] * X[k, :] @ W) < 1):
            J += -Y[k] * X[k, :].T.reshape((-1, 1))
        
    J = J/K
    J = np.reshape(J, (-1, 1))
    #end_solution

    
    return J
```

```python
def ksvm(kernel, X, Y, X_test):
    """
    Write your implementation of Kernel SVM here.

    Arguments:
    kernel : A callable function that takes 2 samples and returns the score
             for them.
    X : Training data matrix
    Y : Training Labels as -1/1
    X_test : Test data matrix

    Returns:
    support_vectors : The learned support vectors
    preds : The predicted labels as -1/1 on the given Test data
    """
    
    #begin_solution
    
    nData, nDims = X.shape

    # Function to minimize
    def obj_alpha(alpha):
        inter_term = 0
        K = alpha.shape[0]
        for i in range(K):
            for j in range(K):
                inter_term += alpha[i] * alpha[j] * Y[i] * Y[j] * kernel(X[i, :], X[j, :])
        return ((1/2)*inter_term - np.sum(alpha)).item()
            
    
    #alpha_0 = np.random.rand(nData).reshape(-1, 1)
    alpha_0 = np.zeros(nData).reshape(-1, 1)
    
    C = 0.1 # C positive
    bnds = tuple([[0, C] for t in range(nData)])
    cons=[{"type": "eq", "fun": lambda alpha: Y.reshape(1, -1) @ alpha}]
    
    res = optimize.minimize(obj_alpha, alpha_0, constraints=cons, bounds=bnds)
    #res = optimize.minimize(obj_alpha, alpha_0, constraints=cons, bounds=bnds, tol=1e-4)
    
    alpha = res.x
    
    nb_of_support_vectors = 0
    support_vectors = np.zeros_like(X)
    for k in range(nData):
        if(alpha[k] != 0): # If alpha != 0, the vector is a support vector
            support_vectors[nb_of_support_vectors, :] = X[k, :]
            nb_of_support_vectors += 1
    support_vectors = support_vectors[:nb_of_support_vectors+1, :]
    
    nDataTest, nDimsTest = X_test.shape
    preds = np.zeros(nDataTest).reshape(-1, 1)
    for k in range(nDataTest):
        inter_sum = 0
        for i in range(nData):
            inter_sum += alpha[i] * Y[i] * kernel(X[i, :], X_test[k, :])
        if(inter_sum == 0):
            preds[k] = 1
        else:
            preds[k] = np.sign(inter_sum)
            
    #end_solution


    return support_vectors, preds
```

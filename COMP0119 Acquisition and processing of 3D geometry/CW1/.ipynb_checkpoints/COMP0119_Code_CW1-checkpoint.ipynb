```python
# Function computing the normal to every surface point (defined in NormalEstimation during Workshop 1)

def PCA_normal_estimation(surface_points, k):
    num_points, _ = surface_points.shape
    tree = KDTree(surface_points)
    _, indices = tree.query(surface_points, k)
    normals = np.zeros([num_points,3])
    curvatures = np.zeros(num_points)
    for point in range(num_points):
        neighbours = surface_points[indices[point],:]
        mean = np.mean(neighbours,axis=0)
        neighbours_adjust = neighbours - mean
        covariance_matrix = np.cov(neighbours_adjust.T)
        eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)
        sorted_values = np.argsort(-eigenvalues)
        eigenvalues = eigenvalues[sorted_values]
        curvature = abs(eigenvalues[2]) / (abs(eigenvalues[0])+abs(eigenvalues[1])+abs(eigenvalues[2]))
        curvatures[point] = curvature
        eigenvectors = eigenvectors[:,sorted_values]
        normal = eigenvectors[:,2]
        nn = np.linalg.norm(normal)
        normal = normal / nn
        normals[point,:] = normal
    return normals, curvatures
```

```python
# Let's define a subfunction to compute the rigid (R, t) transformation in the point-to-point case

def Rt_pt2pt(p, q):
    # Assuming that p and q have the same dimensions
    # p.shape = (nb_pts, 3)
    
    # Following the ICP algorithm seen in class
    p_mean = np.mean(p, axis=0)
    q_mean = np.mean(q, axis=0)
    p_tilde = p - p_mean
    q_tilde = q - q_mean
    
    A = np.transpose(p_tilde) @ q_tilde
    
    U, S, Vt = np.linalg.svd(A, compute_uv=True)
    
    R = np.transpose(Vt) @ np.transpose(U)
    t = p_mean - (q_mean @ R)
    
    return R, t
    
```

```python
def apply_transform_mesh_and_export(mesh_of_interest, transform_matrices, name='name_test', export=True):
    nb_of_transforms = transform_matrices.shape[0]
    our_mesh = mesh_of_interest.copy()
    
    for k in range(nb_of_transforms):
        if((transform_matrices[k]==np.zeros((4,4))).all()):
            break
        # Trimesh applies the transformation matrices in the reverse order, so we have to transpose the rotation matrix
        # Creating the adapted transformation matrix
        inter_transform = np.eye(4)
        inter_transform[:3, :3] = transform_matrices[k, :3, :3].T
        inter_transform[:3, 3] = transform_matrices[k, :3, 3]
        our_mesh.apply_transform(inter_transform)
    
    if(export):
        our_mesh.export(name+'.ply')
        print('done with exportation')
    else:
        print('done without exportation')    
        
    return our_mesh
```

### 1. ICP point-to-point

```python
def ICP_pt2pt(M1, M2, nb_subsampling, nb_of_iterations, need_epsilon=False, epsilon=1e-3):
    M1_c = M1.copy()
    M2_c = M2.copy()
    
    bool_end_it = True
    ssd_track = np.array([])
    #transform_matrices = np.zeros((4, 4, nb_of_iterations))
    transform_matrices = np.zeros((nb_of_iterations, 4, 4))
    # Creating the tree for nearest neighbours (for iteration 0)
    
    
    # First, sub-sampling and aligning
    # We only sub-sample p, and we'll find the nb of q correspondences after
    # We start by defining a random generator of integers (we sample uniformly)
    #rng = np.random.default_rng()
    #sub_indices = rng.choice(p.shape[0], nb_subsampling, replace=False)

    # Subsampling
    p = M1_c.sample(nb_subsampling)
    q = M2_c.sample(nb_subsampling)
    
    tree_q = KDTree(q)
    
    for nb_it in range(nb_of_iterations):
    
        # Let's now look for the nearest neighbour of each subsampled point
        q_corr = np.zeros_like(p)
        for k in range(q_corr.shape[0]):
            _, nn_idx = tree_q.query(p[k,:].reshape(1, -1), 1)
            q_corr[k, :] = q[nn_idx, :]
            
        # Computing R, t
        R, t = Rt_pt2pt(p, q_corr)
        
        # Storing in transform_matrices
        transform_matrices[nb_it, :3, :3] = R
        transform_matrices[nb_it, :3, 3] = t
        transform_matrices[nb_it, 3, 3] = 1
        
        # Applying the transform
        q = (q @ R) + t
        
        # Computing the new tree
        tree_q = KDTree(q)
        
        # Computing the SSD error
        ssd_error = np.sum((p-q)**2)
        
        if((nb_it+1)%5==0):
            print("Iteration "+str(nb_it+1)+": SSD error= "+str(ssd_error))
        
        if((nb_it>0) and (abs(ssd_error-ssd_track[-1])<epsilon) and need_epsilon):
            ssd_track = np.append(ssd_track, ssd_error)
            bool_end_it = False
            print("The algorithm has reached the desired precision.")
            break;
            
        ssd_track = np.append(ssd_track, ssd_error)
            
    if(bool_end_it):
        print("The algorithm has consumed all of its iterations.")
          
    plt.plot(ssd_track)
    plt.title("SSD error over iterations")
    plt.xlabel("Number of iterations")
    plt.ylabel("Sum of Squared Differences")
    plt.savefig('ICP_pt2pt_ssd.png')
    plt.show()
        
    return ssd_track, transform_matrices
```

Let's create a sub-function that will rotate our mesh from a given angle.
We will then create an array of different angles to see the performance of ICP.
We will also modify the ICP function to stop once it stagnates at one error value.
Finally, we still have to check the result correspondences, because we can guess that some glueing can achieve better scores whereas the sticking is completely irrealisable.

```python
def rotation_mesh(mesh_of_interest, angle):
    new_mesh = mesh_of_interest.copy()
    
    # Angle should be in radians
    R = np.array([[np.cos(angle), -np.sin(angle), 0], [np.sin(angle), np.cos(angle), 0], [0, 0, 1]])
    transform_matrix = np.eye(4)
    transform_matrix[:3, :3] = R.T # Need to transpose for trimesh.apple_transform
    
    new_mesh.apply_transform(transform_matrix)
    
    return new_mesh
```

We will often call the "\_r" version of our ICP function, it is exactly the same, but it does not output any text, nor plots, so as to save time and only output useful information.

### 6. Point-to-plane ICP

```python
def ICP_pt2plane(M1, M2, nb_subsampling, nb_of_iterations, k, k_normal, need_epsilon=False, epsilon=1e-3):
    M1_c = M1.copy()
    M2_c = M2.copy()
    
    bool_end_it = True
    ssd_track = np.array([])
    #transform_matrices = np.zeros((4, 4, nb_of_iterations))
    transform_matrices = np.zeros((nb_of_iterations, 4, 4))
    # Creating the tree for nearest neighbours (for iteration 0)
    
    
    # First, sub-sampling and aligning
    # We only sub-sample p, and we'll find the nb of q correspondences after
    # We start by defining a random generator of integers (we sample uniformly)
    #rng = np.random.default_rng()
    #sub_indices = rng.choice(p.shape[0], nb_subsampling, replace=False)

    # Subsampling
    p = M1_c.sample(nb_subsampling)
    q = M2_c.sample(nb_subsampling)
    
    tree_q = KDTree(q)
    
    normals, _ = PCA_normal_estimation(p, k)
    
    for nb_it in range(nb_of_iterations):
    
        # Let's now look for the nearest neighbour of each subsampled point
        q_corr = np.zeros_like(p)
        for k in range(q_corr.shape[0]):
            _, nn_idx = tree_q.query(p[k,:].reshape(1, -1), k_normal)
            nn_idx = np.reshape(nn_idx, -1)
            min_idx = nn_idx[0]
            min_dist = abs(np.dot(p[k, :]-q[nn_idx[0], :], normals[k, :]))
            for l in range(1, k_normal):
                dist_l = abs(np.dot(p[k, :]-q[nn_idx[l], :], normals[k, :]))
                if(dist_l < min_dist):
                    min_idx = nn_idx[l]
                    min_dist = dist_l
            q_corr[k, :] = q[min_idx, :]
                
            
        # Computing R, t
        R, t = Rt_pt2pt(p, q_corr)
        
        # Storing in transform_matrices
        transform_matrices[nb_it, :3, :3] = R
        transform_matrices[nb_it, :3, 3] = t
        transform_matrices[nb_it, 3, 3] = 1
        
        # Applying the transform
        q = (q @ R) + t
        
        # Computing the new tree
        tree_q = KDTree(q)
        
        # Computing the SSD error
        ssd_error = np.sum((p-q)**2)
        
        if((nb_it+1)%5==0):
            print("Iteration "+str(nb_it+1)+": SSD error= "+str(ssd_error))
        
        if((nb_it>0) and (abs(ssd_error-ssd_track[-1])<epsilon) and need_epsilon):
            ssd_track = np.append(ssd_track, ssd_error)
            bool_end_it = False
            print("The algorithm has reached the desired precision.")
            break;
            
        ssd_track = np.append(ssd_track, ssd_error)
            
    if(bool_end_it):
        print("The algorithm has consumed all of its iterations.")
          
    plt.plot(ssd_track)
    plt.title("SSD error over iterations")
    plt.xlabel("Number of iterations")
    plt.ylabel("Sum of Squared Differences")
    plt.savefig('ICP_pt2plane_ssd.png')
    plt.show()
        
    return transform_matrices
```

```python
def add_noise_mesh(mesh, std):
    mesh_copy = mesh.copy()
    
    # We scale the normal noise according to the smallest bounding value for each coordinate
    # /!\ We use the centered mesh
    noise_x = std*np.random.randn(mesh_copy.vertices.shape[0])*max(abs(np.amin(mesh_copy.vertices[:, 0])),abs(np.amax(mesh_copy.vertices[:, 0])))
    noise_y = std*np.random.randn(mesh_copy.vertices.shape[0])*max(abs(np.amin(mesh_copy.vertices[:, 1])),abs(np.amax(mesh_copy.vertices[:, 1])))
    noise_z = std*np.random.randn(mesh_copy.vertices.shape[0])*max(abs(np.amin(mesh_copy.vertices[:, 2])),abs(np.amax(mesh_copy.vertices[:, 2])))

    # Adding noise to each coordinate
    mesh_copy.vertices[:, 0] += noise_x
    mesh_copy.vertices[:, 1] += noise_y
    mesh_copy.vertices[:, 2] += noise_z
    
    return mesh_copy
```

```python
def my_apply_transform(mesh_of_interest, transform_matrices):
    nb_of_transforms = transform_matrices.shape[0]
    our_mesh = mesh_of_interest.copy()
    
    for k in range(nb_of_transforms):
        if((transform_matrices[k]==np.zeros((4,4))).all()):
            break
        # Trimesh applies the transformation matrices in the reverse order, so we have to transpose the rotation matrix
        # Creating the adapted transformation matrix
        inter_transform = np.eye(4)
        inter_transform[:3, :3] = transform_matrices[k, :3, :3].T
        inter_transform[:3, 3] = transform_matrices[k, :3, 3]
        our_mesh.apply_transform(inter_transform)  
        
    return our_mesh
```

```python
def best_alignment_concatenated(M0, M1, M2, M3, M4, M5, nb_subsampling, nb_iterations, epsilon=1e-3):
    M0_c = M0.copy()
    M1_c = M1.copy()
    M2_c = M2.copy()
    M3_c = M3.copy()
    M4_c = M4.copy()
    M5_c = M5.copy()
    
    nb_max_of_permutations = 10
    
    l = np.array(list(permutations(range(6))))
    np.random.shuffle(l)
    list_perm = l[:nb_max_of_permutations,:] # 6! = 720 permutations, we have to limit the number of them
    
    errors_perm = np.array([])
    
    # First, we iterate over the permutation
    for k in range(nb_max_of_permutations):
        
        # We have to reset the meshes array everytime because otherwise, the meshes stay aligned and each permutation will just obtain a better score than the previous one
        meshes = [M0_c, M1_c, M2_c, M3_c, M4_c, M5_c]
        
        perm_k = list_perm[k,:]
        error_single_perm = 0
        
        # And WITHIN each permutation, we iterate over every mesh
        for nb_op in range(5):
            
            if(nb_op==0): # If first iteration
                _, ssd_track_0, transform_matrix_0 = ICP_pt2pt_r(meshes[perm_k[nb_op]], meshes[perm_k[nb_op+1]], nb_subsampling, nb_iterations, epsilon)
                
                error_single_perm += ssd_track_0[-1]
                
                # Attention a applique direct au mesh, faut faire une copie et reload  chq fois
                #apply_transform_mesh_and_export(meshes[perm_k[nb_op+1]], transform_matrix_0, name='', export=False)
                meshes[perm_k[nb_op+1]] = my_apply_transform(meshes[perm_k[nb_op+1]], transform_matrix_0)
                
                res_mesh = trimesh.util.concatenate([meshes[perm_k[nb_op]], meshes[perm_k[nb_op+1]]])
                
            else:
                _, ssd_track, transform_matrix = ICP_pt2pt_r(res_mesh, meshes[perm_k[nb_op+1]], nb_subsampling, nb_iterations, epsilon)
                
                error_single_perm += ssd_track[-1]
                
                # Attention a applique direct au mesh, faut faire une copie et reload  chq fois
                #apply_transform_mesh_and_export(meshes[perm_k[nb_op+1]], transform_matrix, name='', export=False)
                meshes[perm_k[nb_op+1]] = my_apply_transform(meshes[perm_k[nb_op+1]], transform_matrix)
                
                res_mesh = trimesh.util.concatenate([res_mesh, meshes[perm_k[nb_op+1]]])
            
            print("Permutation "+str(k)+" iteration "+str(nb_op)+" done.")
                
        if(k==0):
            best_perm = perm_k
            best_dist = error_single_perm
            
        else:
            if(error_single_perm<best_dist):
                best_perm = perm_k
                best_dist = error_single_perm
        
        errors_perm = np.append(errors_perm, error_single_perm) # Adding the cumulated error
        
    plt.scatter(np.arange(nb_max_of_permutations), errors_perm)
    plt.title("Different cumulated SSD error cumulated for different permutations")
    plt.xlabel("Permutations")
    plt.ylabel("Cumulated Sum of Squared Differences over stitchings")
    plt.savefig('Cumulated_method_ouais_plot.png')
    plt.show()
    
    return best_perm  
```

```python
def global_reg_brute_force_2(M0, M1, M2, M3, M4, M5, nb_subsampling, nb_iterations, epsilon=1e-3):
    M0_c = M0.copy()
    M1_c = M1.copy()
    M2_c = M2.copy()
    M3_c = M3.copy()
    M4_c = M4.copy()
    M5_c = M5.copy()
    
    meshes = [M0_c, M1_c, M2_c, M3_c, M4_c, M5_c]
    
    error_track = np.array([])
    
    nb_epochs = 10
    
    # On peut itrer sur diff epochs en shufflant le tableau mesh pour amliorer la prcision
    
    # We will orientate every scan individually versus a concatenated version of all the others

    for epoch in range(nb_epochs):
        
        random.shuffle(meshes)
    
        error_epoch = 0

        for nb_m in range(6):

            mesh_to_move = meshes[nb_m]

            # Creating the concatenated mesh
            not_created_yet = 1
            for nb_2 in range(6):
                if(nb_2!=nb_m):
                    if(not_created_yet==1):
                        conc_mesh = meshes[nb_2]
                        not_created_yet = 0
                    else:
                        conc_mesh = trimesh.util.concatenate([conc_mesh, meshes[nb_2]])

            _, ssd_track, transform_matrix = ICP_pt2pt_r(conc_mesh, mesh_to_move, nb_subsampling, nb_iterations, epsilon)

            error_epoch += ssd_track[-1]

            # Attention a applique direct au mesh, faut faire une copie et reload  chq fois
            #apply_transform_mesh_and_export(meshes[perm_k[nb_op+1]], transform_matrix_0, name='', export=False)
            meshes[nb_m] = my_apply_transform(meshes[nb_m], transform_matrix)

            # The only difference with the concatenated is here: we just align mesh by mesh instead of sharing the error

            print("Epoch "+str(epoch)+" iteration "+str(nb_m)+" done.")
            
        error_track = np.append(error_track, error_epoch)
        
    meshes[0].export('iter_glob_align_mesh0.ply')
    meshes[1].export('iter_glob_align_mesh1.ply')
    meshes[2].export('iter_glob_align_mesh2.ply')
    meshes[3].export('iter_glob_align_mesh3.ply')
    meshes[4].export('iter_glob_align_mesh4.ply')
    meshes[5].export('iter_glob_align_mesh5.ply')
            
    plt.plot(np.arange(nb_epochs), error_track)
    plt.title("Different cumulated SSD error cumulated for epochs")
    plt.xlabel("Epochs")
    plt.ylabel("Cumulated Sum of Squared Differences over stitchings")
    plt.savefig('ouais_plot_globalign_iteratedover.png')
    plt.show()
        
```

```python
# This version is better!

def ICP_pt2pt_r_M1M2(M1, M2, nb_subsampling, nb_of_iterations, epsilon=1e-3):
    M1_c = M1.copy()
    M2_c = M2.copy()
    
    ssd_track = np.array([])
    #transform_matrices = np.zeros((4, 4, nb_of_iterations))
    transform_matrices = np.zeros((nb_of_iterations, 4, 4))
    # Creating the tree for nearest neighbours (for iteration 0)
    
    
    # First, sub-sampling and aligning
    # We only sub-sample p, and we'll find the nb of q correspondences after
    # We start by defining a random generator of integers (we sample uniformly)
    #rng = np.random.default_rng()
    #sub_indices = rng.choice(p.shape[0], nb_subsampling, replace=False)

    # Subsampling
    p = M1_c.sample(nb_subsampling)
    q = M2_c.sample(nb_subsampling)
    
    tree_q = KDTree(q)
    
    for nb_it in range(nb_of_iterations):
    
        # Let's now look for the nearest neighbour of each subsampled point
        q_corr = np.zeros_like(p)
        for k in range(q_corr.shape[0]):
            _, nn_idx = tree_q.query(p[k,:].reshape(1, -1), 1)
            q_corr[k, :] = q[nn_idx, :]
            
        # Computing R, t
        R, t = Rt_pt2pt(p, q_corr)
        
        # Storing in transform_matrices
        transform_matrices[nb_it, :3, :3] = R
        transform_matrices[nb_it, :3, 3] = t
        transform_matrices[nb_it, 3, 3] = 1
        
        # Applying the transform
        M2_c = my_apply_transform(M2_c, np.array([transform_matrices[nb_it, :, :]]))
        q_ssd = M2_c.sample(40097)
        p_ssd = M1_c.sample(40097)
        
        # Computing the SSD error
        ssd_error = np.sum((p_ssd-q_ssd)**2)
        
        #q = M2_c.sample(15000)
        
        # Computing the new tree
        tree_q = KDTree(q)
        
        if((nb_it>0) and (abs(ssd_error-ssd_track[-1])<epsilon)):
            ssd_track = np.append(ssd_track, ssd_error)
            break;
            
        ssd_track = np.append(ssd_track, ssd_error)
        
    return nb_it, ssd_track, transform_matrices
```
